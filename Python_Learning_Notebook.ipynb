## 5. Lists

```python
fruits=['Apple','Banana','Mango']
for fruit in fruits:
    print(fruit)
print('Length:', len(fruits))
```

## 6. Dictionaries

```python
person={'name':'Sanjay','course':'Python'}
for k,v in person.items():
    print(k,v)
```

## 8. File Handling

## 9. OOP Basics

```python
class Student:
    def __init__(self,name):
        self.name=name
    def show(self):
        print(self.name)

s=Student('Sanjay')
s.show()
```
